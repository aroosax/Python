# Project: Student Management System

This is an intermediate-level Python project using functions, lists, and dictionaries. It allows you to:

*   Add students
*   view students
*   Search for a students
*   Remove a student
*   sort students according to marks
*   Calculate grades

```python
students = []

# Function to calculate grade from percentage
def calculate_grade(percentage):
    if percentage >= 90:
        return "A+"
    elif percentage >= 80 and percentage < 90:
        return "A"
    elif percentage >= 70 and percentage < 80:
        return "B"
    elif percentage >= 60 and percentage < 70:
        return "C"
    elif percentage >= 50 and percentage < 60:
        return "D"
    else:
        return "F"

# Function to add a student
def add_student():
    name = input("Enter student name: ")
    roll_no = input("Enter roll number: ")
    total_marks = float(input("Enter total marks: "))
    obtained_marks = float(input("Enter obtained marks: "))

    percentage = (obtained_marks / total_marks) * 100
    grade = calculate_grade(percentage)

    student = {
        "name": name,
        "roll_no": roll_no,
        "total_marks": total_marks,
        "obtained_marks": obtained_marks,
        "percentage": round(percentage, 2),
        "grade": grade
    }

    students.append(student)
    print("Student added successfully!")

# Function to view all students
def view_students():
    if not students:
        print("No student records found.")
        return
    for idx, student in enumerate(students, start=1):
        print(f"{idx}. Name: {student['name']}, Roll No: {student['roll_no']}, "
              f"Marks: {student['obtained_marks']}/{student['total_marks']}, "
              f"Percentage: {student['percentage']}%, Grade: {student['grade']}")

# Function to search a student by roll number
def search_student():
    keyword = input("Enter roll number to search: ")
    found = False
    for student in students:
        if student["name"] == keyword or student["roll_no"] == keyword:
            print(f"Found: Name: {student['name']}, Roll No: {student['roll_no']}, "
                  f"Marks: {student['obtained_marks']}/{student['total_marks']}, "
                  f"Percentage: {student['percentage']}%, Grade: {student['grade']}")
            found = True
    if not found:
        print("Student not found.")

# Function to sort students by percentage
def sort_students_by_percentage():
    if not students:
        print("No student records found.")
        return

    sorted_students = sorted(students, key=lambda x: x['percentage'], reverse=True)

    print("\nStudents sorted by percentage (high to low):")
    for idx, student in enumerate(sorted_students, start=1):
        print(f"{idx}. Name: {student['name']}, Roll No: {student['roll_no']}, "
              f"Marks: {student['obtained_marks']}/{student['total_marks']}, "
              f"Percentage: {student['percentage']}%, Grade: {student['grade']}")

# Function to remove a student by roll number
def remove_student():
    roll_no = input("Enter roll number of student to remove: ")
    for student in students:
        if student["roll_no"] == roll_no:
            students.remove(student)
            print("Student removed successfully.")
            return
    print("Student not found.")

# Main menu
def main():
    while True:
        print("\n------ Student Management System ------")
        print("1. Add Student")
        print("2. View All Students")
        print("3. Search Student")
        print("4. Remove Student")
        print("5. Sort Students by Percentage")
        print("6. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            add_student()
        elif choice == '2':
            view_students()
        elif choice == '3':
            search_student()
        elif choice == '4':
            remove_student()
        elif choice == '5':
            sort_students_by_percentage()
        elif choice == '6':
            print("Exiting the program.")
            break
        else:
            print("Invalid choice. Please try again.")

# Run the project
main()
```
